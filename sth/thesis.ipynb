```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
names = ['data/forties_field_flow_data.csv', 'data/ekofisk_field_flow_data.csv']
```

```python
df1 = pd.read_csv(names[0], index_col='Test')
df2 = pd.read_csv(names[1], index_col='Test')
df1
```

```python
plt.figure(figsize=(12, 35))

for i, col in enumerate(df1.columns):
    plt.subplot(len(df1.columns), 1, i+1)
    plt.plot(df1[col])
    plt.title(col)
    plt.ylabel(col)
    plt.grid(True)
```
